```python
# Exercise 1:

# Question 1:
# Viết chương trình nhập vào một số thực. Đơn vị tính cho số này là centimet (cm). Hãy in ra số tương đương tính bằng foot (số thực, có 2 số lẻ thập phân) và inch (số thực, có 2 số lẻ thập phân). Với 1 inch = 2.54 cm và 1 foot = 12 inches.

def convert_inch_foot(cm):
    return cm/2.54, cm/2.54/12 # inch, foot
    
# Question 2:
# Viết chương trình nhập vào số giây từ 0 tới 68399, đổi số giây này thành dạng [giờ:phút:giây], mỗi thành phần là một số nguyên có 2 chữ số.
# Ví dụ: 02:11:05
def format_sec(sec):
    return f"[{sec // 3600 }:{(sec % 3600) // 60}:{sec % (3600*60) % 60}]"
print(format_sec(65))

# Question 3:
# Viết chương trình nhập vào 4 số nguyên.
# Hiển thị ra số lớn nhất và nhỏ nhất
# Ví dụ:
# nhập vào 4 số 1,2,3,4 thì kết quả là:
# Số lớn nhất là: 4
# Số nhỏ nhất là: 1

numbers = []
for i in range(4):
    number = int(input("Nhập số nguyên: "))
    numbers.append(number)

max_number = max(numbers)
min_number = min(numbers)

print("Số lớn nhất là:", max_number)
print("Số nhỏ nhất là:", min_number)

# Question 4:
# Nhập vào hai số tự nhiên a và b. Tính hiệu của hai số đó.
# Nếu hiệu số lớn hơn 0 thì in ra dòng chữ [Số thứ nhất lớn hơn số thứ hai].
# Nếu hiệu số nhỏ hơn 0 thì in ra dòng chữ [Số thứ nhất nhỏ hơn số thứ hai].
a = 6
b = 4
difference = a - b
if difference > 0:
    print("Số thứ nhất lớn hơn số thứ hai")
elif difference < 0:
    print("Số thứ nhất nhỏ hơn số thứ hai")
    
    
# Question 5:
# Nhập vào 2 số tự nhiên a và b. Kiểm tra xem a có chia hết cho b không
a = int(input("Nhập số tự nhiên a: "))
b = int(input("Nhập số tự nhiên b: "))

if a % b == 0:
    print("a chia hết cho b")
else:
    print("a không chia hết cho b")

# Question 6:
# Xếp loại sinh viên theo các quy luật dưới đây:
# -	Nếu điểm >= 9.0 - Loại xuất sắc
# -	Nếu 8.0<= điểm < 9.0 - Loại giỏi
# -	Nếu 7.0<= điểm < 8.0 - Loại khá

# -	Nếu 6.0 <= điểm < 7.0 - Loại trung bình khá
# -	Nếu 5.0 <= điểm < 6.0 - Loại trung bình
# -	Nếu điểm < 5.0 - Loại kém
# Với điểm = ( (điểm toán)*2 + điểm lý + điểm hóa )/4
diem_toan = float(input("Nhập điểm toán: "))
diem_ly = float(input("Nhập điểm lý: "))
diem_hoa = float(input("Nhập điểm hóa: "))

diem_trung_binh = (diem_toan * 2 + diem_ly + diem_hoa) / 4

if diem_trung_binh >= 9.0:
    print("Loại xuất sắc")
elif 8.0 <= diem_trung_binh < 9.0:
    print("Loại giỏi")
elif 7.0 <= diem_trung_binh < 8.0:
    print("Loại khá")
elif 6.0 <= diem_trung_binh < 7.0:
    print("Loại trung bình khá")
elif 5.0 <= diem_trung_binh < 6.0:
    print("Loại trung bình")
else:
    print("Loại kém")

diem_toan = float(input("Nhập điểm môn Toán: "))
diem_ly = float(input("Nhập điểm môn Lý: "))
diem_hoa = float(input("Nhập điểm môn Hóa: "))

diem_trung_binh = (diem_toan + diem_ly + diem_hoa) / 3

if diem_trung_binh >= 9.0:
    hoc_luc = "Xuất sắc"
elif 8.0 <= diem_trung_binh < 9.0:
    hoc_luc = "Giỏi"
elif 7.0 <= diem_trung_binh < 8.0:
    hoc_luc = "Khá"
elif 6.0 <= diem_trung_binh < 7.0:
    hoc_luc = "Trung bình khá"
elif 5.0 <= diem_trung_binh < 6.0:
    hoc_luc = "Trung bình"
else:
    hoc_luc = "Kém"

print("Học lực của học sinh là:", hoc_luc)


# Exercise 2:

# Question 1:
# Hãy viết chương trình tính tổng các chữ số của một số nguyên bất kỳ.
# Ví dụ: Số 8545604 có tổng các chữ số là: 8+5+4+5+6+0+4= 32

number = int(input("Nhập số nguyên: "))
sum_of_digits = 0

while number > 0:
    digit = number % 10
    sum_of_digits += digit
    number //= 10

print("Tổng các chữ số của số nguyên là:", sum_of_digits)


# Question 2:
# Viết chương trình phân tích một số nguyên thành các thừa số nguyên tố
# Ví dụ: Số 28 được phân tích thành 2 x 2 x 7

number = int(input("Nhập số nguyên: "))
factors = []

for i in range(2, number+1):
    while number % i == 0:
        factors.append(i)
        number //= i

if len(factors) == 0:
    factors.append(number)

print("Phân tích số nguyên thành các thừa số nguyên tố:", factors)


# Question 3:
# Viết chương trình liệt kê tất cả các số nguyên tố nhỏ hơn n cho trước.

n = int(input("Nhập số nguyên n: "))

prime_numbers = []

for num in range(2, n):
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        prime_numbers.append(num)

print("Các số nguyên tố nhỏ hơn", n, "là:", prime_numbers)


# Question 4:
# Viết chương trình liệt kê n số nguyên tố đầu tiên.

n = int(input("Nhập số nguyên n: "))

prime_numbers = []
num = 2

while len(prime_numbers) < n:
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        prime_numbers.append(num)
    num += 1

print(n, "số nguyên tố đầu tiên là:", prime_numbers)


# Question 5:
# Viết chương trình tìm ước số chung lớn nhất, bội số chung nhỏ nhất của hai số tự nhiên a và b.

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def lcm(a, b):
    return (a * b) // gcd(a, b)

a = int(input("Nhập số tự nhiên a: "))
b = int(input("Nhập số tự nhiên b: "))

gcd_result = gcd(a, b)
lcm_result = lcm(a, b)

print("Ước số chung lớn nhất của", a, "và", b, "là:", gcd_result)
print("Bội số chung nhỏ nhất của", a, "và", b, "là:", lcm_result)


# Question 6:
# Dãy số Fibonacci được định nghĩa như sau:
# F0 =1, F1 = 1; Fn = Fn-1 + Fn-2 với n>=2.
# Hãy viết chương trình tìm số Fibonacci thứ n.

n = int(input("Nhập số nguyên n: "))

fibonacci_sequence = [1, 1]

for i in range(2, n):
    fibonacci_sequence.append(fibonacci_sequence[i-1] + fibonacci_sequence[i-2])

fibonacci_number = fibonacci_sequence[n-1]

print("Số Fibonacci thứ", n, "là:", fibonacci_number)


# Question 7:
# Nhập một số tự nhiên n.
# Hãy liệt kê các số Fibonaci nhỏ hơn n là số nguyên tố

n = int(input("Nhập số tự nhiên n: "))

factorial = 1
for i in range(1, n+1):
    factorial *= i

print(n, "! =", factorial)
```
