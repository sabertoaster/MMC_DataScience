Vi bi ny, ta cn xy dng nng cp chc nng c/ghi file v t chc xy dng thnh mt module.

Ta s cn xy dng trc mt hm c/ghi file  c th ti s dng nhiu ln. Vi u vo u ra nh sau:

|Hm|u vo|u ra|
|--|--|--|
|doc_file|ng dn ti file, mode c file|D liu dng list c c t file|
|ghi_file|D liu ghi file, tn ca file v mode ghi file|Trng thi nu c li ghi file|


````python
import random
````

````python
def ghi_file(data, ten_file, mode):
    try:
        with open(ten_file, mode) as file:
            for line in data:
                file.write(line)
    except Exception as e:
        print(f"C li xy ra: {e}")


def doc_file(duong_dan, mode):
    try:
        with open(duong_dan, mode) as file:
            du_lieu = [line.strip().split(',') for line in file.readlines()]
        return du_lieu
    except Exception as e:
        print(f"C li xy ra khi c file: {e}")
        return None
````

Tip tc, to hm kim tra ti khon vi u vo l ti khon mt khu ngi dng, kt hp vi hm kim tra admin. Tin hnh ly danh sch ti khon, s dng hm c file  c. Nu tr v 0,1 th c coi l ng nhp thnh cng, tr v null (khng c gi tr g) th l ng nhp tht bi:

````python
def kiem_tra_tai_khoan(username, password):
    global dang_nhap_status, user
    # c d liu t file ti khon
    lst_taikhoan = doc_file(PATH_DATA_TAI_KHOAN, "r")
    if lst_taikhoan is None:
        return

    for tai_khoan in lst_taikhoan:
        if tai_khoan[0] == username and tai_khoan[1] == password:
            return kiem_tra_admin(username)
````

Ta to hm kim tra ng nhp, nu true l ng nhp thnh cng, false l tht bi

````python
def check_dang_nhap(username, password):
    if kiem_tra_tai_khoan(username, password) in (0,1):
        return True
    else:
        return False
````

````python
def dang_nhap():
    global dang_nhap_status, user
    print("---GAME L  HC---")
    print("Vui lng ng nhp  chi game")

    while True:
        username = input("Nhp username: ")
        password = input("Nhp password: ")

        if check_dang_nhap(username, password):
            user = username
            dang_nhap_status = True
            break
        else:
            print("Sai tn ngi dng hoc mt khu, vui lng th li.")
````

#### Chc nng np tin

Ta xc nh u vo s l username v s tin cn np. Sau  ta ly thng tin t list ti khon  c c cng vi s tin np sau  lu li l thnh cng

````python
def nap_tien(username, so_tien_nap):
    # c d liu t file ti khon
    lst_taikhoan = doc_file(PATH_DATA_TAI_KHOAN, "r")

    # Tm ti khon trong danh sch
    for tai_khoan_info in lst_taikhoan:
        if tai_khoan_info[0] == username:
            # Cng s tin np vo s tin ca ti khon
            tai_khoan_info[2] = str(int(tai_khoan_info[2]) + so_tien_nap)
            # Ghi li d liu mi vo file
            ghi_file([','.join(tai_khoan) + '\n' for tai_khoan in lst_taikhoan], PATH_DATA_TAI_KHOAN, "w")
            print(f" np {so_tien_nap} vo ti khon {username}.")
            return

    print("Khng tm thy ti khon.")
    return
````

#### Chc nng chi l

Nh phin bn trc, ta gi nguyn cc hm quay s, in danh sch gii

````python
def quay_so():
    danh_sach_giai = {}

    ten_giai = ["GII NHT", "GII NH", "GII BA", "GII T", "GII NM", "GII SU", "GII BY"]

    for giai in ten_giai:
        so = ""
        for i in range(5):
            so += str(random.randint(0, 9))  # To mt ch s ngu nhin t 0 n 9 v thm vo s
        danh_sach_giai[giai] = so

    return danh_sach_giai

def in_danh_sach_giai(danh_sach_giai):
    print('-' * 20)
    print("---DANH SCH GII---")
    for giai, so in danh_sach_giai.items():
        print(f"{giai}: {so}")
    print('-' * 20)
````

Thng tin  c xut ra mn hnh nn ta ch cn to hm lu file thng tin  vo file choilo.txt l thnh cng.Ta s lu cc thng tin `username, danh_sach_so_lo, tien_cuoc, danh_sach_giai, tien_thang, tien_thua`

Ta import th vin datetime  lu thi gian chi

````python
from datetime import datetime
````

````python
def luu_thong_tin_choi_lo(username, danh_sach_so_lo, tien_cuoc, danh_sach_giai,  tien_thang, tien_thua):
    try:
        thoi_gian_choi = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        thong_tin = f"{thoi_gian_choi},{username},{','.join(danh_sach_so_lo)},{tien_cuoc}," \
                    f"{','.join(danh_sach_giai.values())},{tien_thang},{tien_thua}\n"

        ghi_file(thong_tin, PATH_DATA_CHOI_LO, "a")
        print("Thng tin v ln chi l  c lu.")

        # Cp nht s tin thng/thua vo ti khon

    except Exception as e:
        print(f"C li xy ra khi lu thng tin: {e}")
````

Ngoi ra khi chi l, tin l trong file ti khon cng b thay i, nn ta cn cp nht tin chi l c vo file ti khon

````python
def cap_nhat_tien_choi_lo(username, tong_tien):
    try:
        du_lieu_tai_khoan = doc_file(PATH_DATA_TAI_KHOAN, "r")
        if du_lieu_tai_khoan:
            for i, tai_khoan in enumerate(du_lieu_tai_khoan):
                if tai_khoan[0] == username:
                    du_lieu_tai_khoan[i][2] = str(tong_tien)  # Cp nht s tin
                    ghi_file([','.join(tai_khoan) + '\n' for tai_khoan in du_lieu_tai_khoan], PATH_DATA_TAI_KHOAN, "w")
                    return
    except Exception as e:
        print(f"C li xy ra khi cp nht thng tin ti khon: {e}")
````

Ghp 2 hm trn vo hm game quay s l xong

````python
def game_quay_so(username, danh_sach_so_lo, tien_cuoc):
    tong_tien = int(lay_thong_tin_tai_khoan(username)[2])

    danh_sach_giai = quay_so()

    in_danh_sach_giai(danh_sach_giai)
    dem_so_lo_trung = 0
    so_lo_trung = []  # Danh sch lu tr s l trng

    print("Cc s l bn chi l:", danh_sach_so_lo)
    for giai, so_giai in danh_sach_giai.items():
        for so_lo in danh_sach_so_lo:
            if so_lo == so_giai[-2:]:
                dem_so_lo_trung += 1
                so_lo_trung.append(so_lo)
    tien_thang = 0  # Khi to tien_thang
    tien_thua = 0   # Khi to tien_thua

    if dem_so_lo_trung > 0:
        tien_thang = tien_cuoc * dem_so_lo_trung * 70
        tien_thua = tien_cuoc * (len(danh_sach_so_lo) - dem_so_lo_trung)

        tong_tien += tien_thang - tien_thua

        print(f"Bn  trng {dem_so_lo_trung} nhy!")
        print("Cc s l trng:", ", ".join(so_lo_trung))
        print(f"S tin bn trng l {tien_thang}")
        print(f"S tin bn thua l {tien_thua}")
        print(f"Tng tin bn c l {tong_tien}")
    else:
        tien_thua = tien_cuoc * len(danh_sach_so_lo)
        tong_tien -= tien_thua

        print("Bn  thua l!")
        print(f"S tin bn thua l {tien_thua}")
        print(f"Tng tin bn cn l {tong_tien}")

    # Lu thng tin v ln chi l vo file
    luu_thong_tin_choi_lo(username, danh_sach_so_lo, tien_cuoc, danh_sach_giai, tien_thang, tien_thua)
    cap_nhat_tien_choi_lo(username, tong_tien)
````

Hon thin hm thng k

````python
def thong_ke_user(username):
    try:
        du_lieu_choi_lo = doc_file(PATH_DATA_CHOI_LO, "r")
        so_lan_choi = 0
        tong_tien_thang = 0
        tong_tien_thua = 0
        so_lan_thang = 0
        so_lan_thua = 0

        for choi_lo in du_lieu_choi_lo:
            if choi_lo[1] == username:
                so_lan_choi += 1
                tong_tien_thang += int(choi_lo[-2])
                tong_tien_thua += int(choi_lo[-1])
                if int(choi_lo[-2]) >  0:
                    so_lan_thang += 1
                else:
                    so_lan_thua += 1
        ti_le_thang = round(so_lan_thang / so_lan_thua, 2) if so_lan_thua > 0 else 0

        # In kt qu
        print("===THNG K CHI L USER: "+ username+"===")
        print("S ln chi l:", so_lan_choi)
        print("Tng tin chi l thng:", tong_tien_thang)
        print("Tng tin chi l thua:", tong_tien_thua)
        print("T l chi l thng:", ti_le_thang)

    except Exception as e:
        print(f"C li xy ra khi thc hin thng k: {e}")
````

#### Thng k mc admin

Ta lm tng t nhng khng cn chy vng lp kim tra username

````python
def thong_ke_admin():
    try:
        # c d liu t file choilo.txt
        du_lieu_choi_lo = doc_file(PATH_DATA_CHOI_LO, "r")
        so_luong_tai_khoan = len(doc_file(PATH_DATA_TAI_KHOAN, "r"))
        tong_so_lan_choi_lo = len(du_lieu_choi_lo)
        tong_tien_thang = sum(int(choi_lo[11]) for choi_lo in du_lieu_choi_lo)
        tong_tien_thua = sum(int(choi_lo[12]) for choi_lo in du_lieu_choi_lo)
        so_lan_thang = sum(1 for choi_lo in du_lieu_choi_lo if int(choi_lo[-2]) > 0)
        so_lan_thua = tong_so_lan_choi_lo - so_lan_thang

        # Tnh t l thng/thua
        ti_le_thang_thua = round(so_lan_thang / so_lan_thua, 2) if so_lan_thua > 0 else 0

        # In kt qu thng k
        print("===THNG K TNG HP===")
        print("S lng ti khon:", so_luong_tai_khoan)
        print("Tng s lt chi l:", tong_so_lan_choi_lo)
        print("Tng tin chi l thng:", tong_tien_thang)
        print("Tng tin chi l thua:", tong_tien_thua)
        print("T l thng/thua:", ti_le_thang_thua)

    except Exception as e:
        print(f"C li xy ra khi thc hin thng k tng hp: {e}")
````

def ghi_file(data, ten_file, mode):
    try:
        with open(ten_file, mode) as file:
            for line in data:
                file.write(line)
    except Exception as e:
        print(f"C li xy ra: {e}")


def doc_file(duong_dan, mode):
    try:
        with open(duong_dan, mode) as file:
            du_lieu = [line.strip().split(',') for line in file.readlines()]
        return du_lieu
    except Exception as e:
        print(f"C li xy ra khi c file: {e}")
        return None

import random
from .XuLyTaiKhoan import *
from datetime import datetime

def quay_so():
    danh_sach_giai = {}

    ten_giai = ["GII NHT", "GII NH", "GII BA", "GII T", "GII NM", "GII SU", "GII BY"]

    for giai in ten_giai:
        so = ""
        for i in range(5):
            so += str(random.randint(0, 9))  # To mt ch s ngu nhin t 0 n 9 v thm vo s
        danh_sach_giai[giai] = so

    return danh_sach_giai

def in_danh_sach_giai(danh_sach_giai):
    print('-' * 20)
    print("---DANH SCH GII---")
    for giai, so in danh_sach_giai.items():
        print(f"{giai}: {so}")
    print('-' * 20)

def nhap_so_lo():
    while True:
        so_lo_input = input("Nhp cc s l bn mun chi, cch nhau bng du phy: ")
        danh_sach_so_lo = so_lo_input.split(",")

        so_lo_hop_le = True

        for so_lo in danh_sach_so_lo:
            if len(so_lo) != 2 or not so_lo.isdigit():
                print("S l nhp khng hp l. Vui lng nhp li.")
                so_lo_hop_le = False
                break

        if so_lo_hop_le:
            return danh_sach_so_lo
def nhap_tien_cuoc(username,so_lo):
    tong_tien = int(lay_thong_tin_tai_khoan(username)[2])
    while True:
        tien_cuoc = input("Nhp tin cc m bn mun cc (tin cc < tng tin bn c): ")
        if tien_cuoc.isdigit() and int(tien_cuoc) * so_lo < tong_tien:
            return int(tien_cuoc)
        else:
            print("Nhp sai nh dng hoc qu tng tin. Nhp li")


def cap_nhat_tien_choi_lo(username, tong_tien):
    try:
        du_lieu_tai_khoan = doc_file(PATH_DATA_TAI_KHOAN, "r")
        if du_lieu_tai_khoan:
            for i, tai_khoan in enumerate(du_lieu_tai_khoan):
                if tai_khoan[0] == username:
                    du_lieu_tai_khoan[i][2] = str(tong_tien)  # Cp nht s tin
                    ghi_file([','.join(tai_khoan) + '\n' for tai_khoan in du_lieu_tai_khoan], PATH_DATA_TAI_KHOAN, "w")
                    return
    except Exception as e:
        print(f"C li xy ra khi cp nht thng tin ti khon: {e}")


def luu_thong_tin_choi_lo(username, danh_sach_so_lo, tien_cuoc, danh_sach_giai,  tien_thang, tien_thua):
    try:
        thoi_gian_choi = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        thong_tin = f"{thoi_gian_choi},{username},{','.join(danh_sach_so_lo)},{tien_cuoc}," \
                    f"{','.join(danh_sach_giai.values())},{tien_thang},{tien_thua}\n"
        
        ghi_file(thong_tin, PATH_DATA_CHOI_LO, "a")
        print("Thng tin v ln chi l  c lu.")
        
        # Cp nht s tin thng/thua vo ti khon

    except Exception as e:
        print(f"C li xy ra khi lu thng tin: {e}")

def game_quay_so(username, danh_sach_so_lo, tien_cuoc):
    tong_tien = int(lay_thong_tin_tai_khoan(username)[2])

    danh_sach_giai = quay_so()
    
    in_danh_sach_giai(danh_sach_giai)
    dem_so_lo_trung = 0
    so_lo_trung = []  # Danh sch lu tr s l trng

    print("Cc s l bn chi l:", danh_sach_so_lo)
    for giai, so_giai in danh_sach_giai.items():
        for so_lo in danh_sach_so_lo:
            if so_lo == so_giai[-2:]:
                dem_so_lo_trung += 1
                so_lo_trung.append(so_lo)
    tien_thang = 0  # Khi to tien_thang
    tien_thua = 0   # Khi to tien_thua

    if dem_so_lo_trung > 0:
        tien_thang = tien_cuoc * dem_so_lo_trung * 70
        tien_thua = tien_cuoc * (len(danh_sach_so_lo) - dem_so_lo_trung)

        tong_tien += tien_thang - tien_thua

        print(f"Bn  trng {dem_so_lo_trung} nhy!")
        print("Cc s l trng:", ", ".join(so_lo_trung))
        print(f"S tin bn trng l {tien_thang}")
        print(f"S tin bn thua l {tien_thua}")
        print(f"Tng tin bn c l {tong_tien}")
    else:
        tien_thua = tien_cuoc * len(danh_sach_so_lo)
        tong_tien -= tien_thua

        print("Bn  thua l!")
        print(f"S tin bn thua l {tien_thua}")
        print(f"Tng tin bn cn l {tong_tien}")

    # Lu thng tin v ln chi l vo file
    luu_thong_tin_choi_lo(username, danh_sach_so_lo, tien_cuoc, danh_sach_giai, tien_thang, tien_thua)
    cap_nhat_tien_choi_lo(username, tong_tien)
